```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def FuncionC(x):
    return np.sqrt(np.tan(x))
```

```python
def DerivadaExacta(x):
    return (1/(np.cos(x)**2))/(2*np.sqrt(np.tan(x)))
```

```python
def derivadaOrden2PuntoC(funcion, x, h=0.01):
    d = 0
    if h != 0:
        d = (1/(2*h))*((-3)*funcion(x) + (4*funcion(x+h)) - (funcion(x+2*h)))
    return d
```

```python
def derivadaCentrarPuntoD(funcion, x, h=0.01):
    d = 0
    if h != 0:
        d = (funcion(x+h)-funcion(x-h))/(2*h)
    return d
```

```python
x = np.linspace(0.1, 1.1, 100)
C = derivadaOrden2PuntoC(FuncionC, x)
D = derivadaCentrarPuntoD(FuncionC, x)
Ex = DerivadaExacta(x)
```

```python
ErrorNodalDerProgresiva = np.abs(Ex-C)
ErrorNodalDerCentral = np.abs(D-Ex)
```

```python
plt.plot(x, D)
plt.title("Derivada de segundo orden")
```

```python
plt.plot(x, C)
plt.title("Derivada Central")
```

```python
plt.plot(x, ErrorNodalDerCentral)
plt.title("Error de la derivada Central")
```

```python
plt.plot(x, ErrorNodalDerProgresiva)
plt.title("Error de la derivada de segundo orden")
```
